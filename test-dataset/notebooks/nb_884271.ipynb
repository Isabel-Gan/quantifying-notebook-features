```python
import numpy as np
from itertools import groupby
import pandas as pd
import locale as lo
from collections import Counter

def distance(point1, point2):

	deltaX = float(point2[0]) - float(point1[0])
	deltaY = float(point2[1]) - float(point1[1])

	return np.sqrt(deltaX*deltaX + deltaY*deltaY)


def binding(buszok,tornyok):
	asd = []
	for i,busz in buszok.iterrows():
		distances = []
		for j,torony in tornyok.iterrows():

			dist = distance([torony["latitude"],torony["longitude"]],[busz["lat"],busz["long"]])
			distances.append(dist)
		
		legkozelebb = distances.index(min(distances))
		
		sting = str(tornyok.ix[legkozelebb][0]) + " "
		sting = sting + str(tornyok.ix[legkozelebb][1]) + ","		
		asd.append(sting)
		
	
		print(".")
		
	return Counter(asd)

def makeCluster(data):
	return [len(list(group)) for key, group in groupby(data)]


bus_ = r'ee.txt'
busstop = pd.read_csv(bus_,names = ["lat","long"])	

full = r'fulldata.csv'
fulldata = pd.read_csv(full,sep=";")

tornyok = pd.DataFrame()
tornyok["latitude"] = fulldata["latitude"]
tornyok["longitude"] = fulldata["longitude"]

data = binding(busstop,tornyok)

with open('./binding.txt', 'a') as f1:
	f1.write(data)


print(data)
```
