```python
from bs4 import BeautifulSoup
from datetime import datetime
import newspaper
import requests
import json
import os

##pull html
#url = "http://itmakessenseblog.com/"
#content = urllib.request.urlopen(url).read()
#soup = BeautifulSoup(content)
#print(soup.prettify())

#reading html flat files 
source = 'usanewsflash'
month = '02'
days = [str(i) for i in list(range(18,28))]

dates = []
for day in days:
    dates.append("{0}-{1}".format(month, day))
    
links = []   
for date in dates:
    filepath = 'data/notCredible/{0}/{1}.html'.format(source,date)
    with open(filepath) as f:
        soup = BeautifulSoup(f, 'lxml')
        
    # Pull out article links

    i=0
    
    for html_link in soup.find_all('a', {'rel': 'bookmark'}):
        url = html_link.get('href')
        if url not in links:
            links.append(url)

for link in links:        
    link = newspaper.Article(link)
    link.download()
    html = link.html
    link.parse()
    date_longform = ''
    title = link.title.replace(' - USA News Zone', '')

    article = {}
    article["html"] = html
    article["title"] = title
    article["url"] = link.url
    article["date"] = date_longform
    article["source"] = source
    article["text"] = link.text
    article["images"] = list(link.images)
    article["videos"] = link.movies

    outfile = 'data/notCredible/{0}/articles/article{1}.txt'.format(source,i)
    os.makedirs(os.path.dirname(outfile), exist_ok=True)
    i+=1

    with open(outfile, 'w') as f:
        json.dump(article, f)
```
